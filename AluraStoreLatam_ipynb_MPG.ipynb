### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

# 2. Ventas por categoría

```python
Venta_categoria1 = tienda.groupby('Categoría del Producto')['Producto'].count().reset_index(name="Ventas")
Venta_categoria1
```

```python
Venta_categoria2 = tienda2.groupby('Categoría del Producto')['Producto'].count().reset_index(name="Ventas")
Venta_categoria2
```

```python
Venta_categoria3 = tienda3.groupby('Categoría del Producto')['Producto'].count().reset_index(name="Ventas")
Venta_categoria3
```

```python
Venta_categoria4 = tienda4.groupby('Categoría del Producto')['Producto'].count().reset_index(name="Ventas")
Venta_categoria4
```

```python
ventas_totales_categoria = pd.concat([
    Venta_categoria1,
    Venta_categoria2,
    Venta_categoria3,
    Venta_categoria4
]).groupby('Categoría del Producto')['Ventas'].sum().reset_index()

plt.figure(figsize=(10, 8))
plt.pie(ventas_totales_categoria['Ventas'], labels=ventas_totales_categoria['Categoría del Producto'], autopct='%1.1f%%', startangle=140)
plt.title('Ventas Totales por Categoría de Producto en Todas las Tiendas')
plt.axis('equal')
plt.show()
```

# 3. Calificación promedio de la tienda


```python
promedio1 = tienda['Calificación'].mean()
print(f"Tienda 1: {promedio1:.2f}")
```

```python
promedio2 = tienda2['Calificación'].mean()
print(f"Tienda 2: {promedio2:.2f}")
```

```python
promedio3 = tienda3['Calificación'].mean()
print(f"Tienda 3: {promedio3:.2f}")
```

```python
promedio4 = tienda4['Calificación'].mean()
print(f"Tienda 4: {promedio4:.2f}")
```

```python
import matplotlib.pyplot as plt

promedio = [promedio1, promedio2, promedio3, promedio4]
tienda = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
data = {'Tienda': tienda, 'Promedio': promedio}
df = pd.DataFrame(data)

ax = df.plot(kind='line', x='Tienda', y='Promedio', marker='o')


for i, txt in enumerate(df['Promedio']):
    ax.annotate(f'{txt:.2f}', (i, txt), textcoords="offset points", xytext=(0,10), ha='center')

plt.title('Calificación Promedio por Tienda')
plt.xlabel('Tiendas')
plt.ylabel('Promedio de Calificación')
plt.grid(True)
plt.show()
```

La tienda con el mejor promedio de calificacion es la Tienda 3

# 4. Productos más y menos vendidos

```python
tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tiendas = [tienda1, tienda2, tienda3, tienda4]
```

```python
import pandas as pd

tiendas_total = pd.concat(tiendas, ignore_index=True)
```

```python
mas_vendidos_total = tiendas_total['Producto'].value_counts().head(5)
mas_vendidos_total
```

```python
menos_vendidos_total = tiendas_total['Producto'].value_counts().tail(5)
menos_vendidos_total
```

```python
plt.figure(figsize=(10,6))

mas_vendidos_total.plot(kind='bar', color='skyblue')

plt.title('Top 5 Productos Más Vendidos (Global)')
plt.xlabel('Producto')
plt.ylabel('Cantidad de Ventas')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

for i, valor in enumerate(mas_vendidos_total.values):
    plt.text(i, valor, str(valor), ha='center', va='bottom')


plt.show()
```

```python
menos_vendidos_total = tiendas_total['Producto'].value_counts().tail(5)

plt.figure(figsize=(10,6))
menos_vendidos_total.plot(kind='bar', color='salmon')

plt.title('Top 5 Productos Menos Vendidos (Global)')
plt.xlabel('Producto')
plt.ylabel('Cantidad de Ventas')
plt.xticks(rotation=45)

for i, valor in enumerate(menos_vendidos_total.values):
    plt.text(i, valor, str(valor), ha='center', va='bottom')

plt.show()
```

# 5. Envío promedio por tienda

```python
envio_prom_t1 = tienda1['Costo de envío'].mean()
envio_prom_t2 = tienda2['Costo de envío'].mean()
envio_prom_t3 = tienda3['Costo de envío'].mean()
envio_prom_t4 = tienda4['Costo de envío'].mean()

print("Tienda 1:", envio_prom_t1)
print("Tienda 2:", envio_prom_t2)
print("Tienda 3:", envio_prom_t3)
print("Tienda 4:", envio_prom_t4)
```

```python
import matplotlib.pyplot as plt

promedios = []

for tienda in tiendas:
    promedios.append(tienda['Costo de envío'].mean())

plt.figure(figsize=(8,5))
plt.bar(['Tienda 1','Tienda 2','Tienda 3','Tienda 4'], promedios)

plt.title('Envío Promedio por Tienda')
plt.ylabel('Costo Promedio de Envío')

for i, valor in enumerate(promedios):
    plt.text(i, valor, f"{valor:.2f}", ha='center', va='bottom')

plt.show()
```
